# Lematizador con Diccionario precargado

```python
class MyLematizador2:
    def __init__(self, language='es', diccionario_path='spanish.txt.learn.flat'):
        self.language = language
        self.diccionario = self.cargar_diccionario(diccionario_path)

    def cargar_diccionario(self, path):
        diccionario = {}
        try:
            with open(path, 'r', encoding='utf-8') as archivo:
                for linea in archivo:
                    partes = linea.strip().split('\t')
                    if len(partes) >= 3:
                        palabra, lema, categoria = partes
                        diccionario[palabra] = lema
        except FileNotFoundError:
            print(f"Error: El archivo {path} no se encuentra.")
        except Exception as e:
            print(f"Error al cargar el diccionario: {e}")
        return diccionario

    def normalizar(self, word=''):
        word = word.lower()
        caracteres_validos = []

        for char in word:
            if char.isalpha() or char.isspace():
                caracteres_validos.append(char)

        palabra_normalizada = ''.join(caracteres_validos)
        return palabra_normalizada

    def lematize(self, text=''):
        text = self.normalizar(text)
        palabras = text.split()

        lematizado = [self.diccionario.get(palabra, palabra) for palabra in palabras]

        return ' '.join(lematizado)


lematizador = MyLematizador2()

texto1 = '¡Viva México, el mejor país del mundo. El maíz es de México!'
print(lematizador.lematize(texto1))

texto2 = '¡Viva México, 100 veces, y más #súper!!!'
print(lematizador.lematize(texto2))

texto3 = 'Viva México'
print(lematizador.lematize(texto3))
```
