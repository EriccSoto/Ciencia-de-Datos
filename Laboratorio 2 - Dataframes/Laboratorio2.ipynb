Transformar datos para otros indicadores

```python
import random
def calcularGanancias(retweets):
  ganancia= retweets * random.randint(3,5)
  return ganancia

```

Agregamos nueva columna con nuevos valores usando la funcion que se cre√≥ usando de base los valores de la columna retweets

```python
df["Ganancias"] = df ["retweets"].apply(calcularGanancias)
```

```python
df.head(10)
```

Calcular una nueva columna a partir de no una sino de varias columnas

```python
def popularidad(fila):
  resultado = fila["followees"]/fila["followers"]
  return resultado
```

Aplicamos la funci√≥n:

**axis=1** indica que se debe aplicar a todas las columnas
**axis=0** es para las filas

```python
df["Popularidad"] = df.apply(popularidad, axis=1)
```
